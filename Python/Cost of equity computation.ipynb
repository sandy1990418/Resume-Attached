```python
import pandas as pd 
import numpy as np 
import os 
from scipy.optimize import fsolve
from IPython.display import display
from scipy import optimize
```

```python
df = pd.read_stata("20210224_v2.dta")
```

## Cost of equity capital GLS Model ---- without imputation

```python
nrow = df.shape[0]
COE= [] 
F_COE =[]
for i in range(0,nrow):
    def f(x):
        return (
            df['bkvlps'][i]+
            (((df['FROE_1'][i]-x)*df['bkvlps'][i])/(1+x))+
            (((df['FROE_2'][i]-x)*df['FBPS_1'][i])/((1+x)**2))+
            (((df['FROE_3'][i]-x)*df['FBPS_2'][i])/((1+x)**3))+
            (((df['FROE_4'][i]-x)*df['FBPS_3'][i])/((1+x)**4))+
            (((df['FROE_5'][i]-x)*df['FBPS_4'][i])/((1+x)**5))+
            (((df['FROE_6'][i]-x)*df['FBPS_5'][i])/((1+x)**6))+
            (((df['FROE_7'][i]-x)*df['FBPS_6'][i])/((1+x)**7))+
            (((df['FROE_8'][i]-x)*df['FBPS_7'][i])/((1+x)**8))+
            (((df['FROE_9'][i]-x)*df['FBPS_8'][i])/((1+x)**9))+
            (((df['FROE_10'][i]-x)*df['FBPS_9'][i])/((1+x)**10))+
            (((df['FROE_11'][i]-x)*df['FBPS_10'][i])/((1+x)**11))+
            (((df['ind_mid_roe'][i]-x)*df['FBPS_11'][i])/(((1+x)**11)*x))-df['prcc_f'][i])
    starting_guess = 0.00000000000007
    coe = optimize.fsolve(f, starting_guess)
    COE.append(coe[0])
    F_COE.append(f(coe[0]))
    print("現在跑號碼:",i,"\n實際數值:",coe,"\n function value:",f(coe[0])) 
df['R_Gebhardt']=COE
df['R_Gebhardt_function_value']=F_COE
```

```python
DES = df.groupby("fyear")['R_Gebhardt'].describe().T
display(DES)
```

## Cost of equity capital GLS Model ---- with imputation

```python
nrow = df.shape[0]
COE= [] 
F_COE =[]
for i in range(0,nrow):
    def f_im(x):
        return (
            df['bkvlps'][i]+
            (((df['FROE_1_im'][i]-x)*df['bkvlps'][i])/(1+x))+
            (((df['FROE_2_im'][i]-x)*df['FBPS_1_im'][i])/((1+x)**2))+
            (((df['FROE_3_im'][i]-x)*df['FBPS_2_im'][i])/((1+x)**3))+
            (((df['FROE_4_im'][i]-x)*df['FBPS_3_im'][i])/((1+x)**4))+
            (((df['FROE_5_im'][i]-x)*df['FBPS_4_im'][i])/((1+x)**5))+
            (((df['FROE_6_im'][i]-x)*df['FBPS_5_im'][i])/((1+x)**6))+
            (((df['FROE_7_im'][i]-x)*df['FBPS_6_im'][i])/((1+x)**7))+
            (((df['FROE_8_im'][i]-x)*df['FBPS_7_im'][i])/((1+x)**8))+
            (((df['FROE_9_im'][i]-x)*df['FBPS_8_im'][i])/((1+x)**9))+
            (((df['FROE_10_im'][i]-x)*df['FBPS_9_im'][i])/((1+x)**10))+
            (((df['FROE_11_im'][i]-x)*df['FBPS_10_im'][i])/((1+x)**11))+
            (((df['ind_mid_roe'][i]-x)*df['FBPS_11_im'][i])/(((1+x)**11)*x))-df['prcc_f_im'][i])
    starting_guess = 0.00000000000007
    coe = optimize.fsolve(f_im, starting_guess)
    COE.append(coe[0])
    F_COE.append(f(coe[0]))
    print("現在跑號碼:",i,"\n實際數值:",coe,"\n function value:",f_im(coe[0])) 
df['R_Gebhardt_im']=COE
df['R_Gebhardt_function_value_im']=F_COE
```

```python
DES = df.groupby("fyear")['R_Gebhardt_im'].describe().T
display(DES)
```

## Cost of equity  MPEG(Easton) model ---- without imputaiton

```python
COE = [] 
nrow = df.shape[0]
F_COE =[]

for i in range(0,nrow):
    def  f_easton(x):
        return (
            (x*df['FDPS_1'][i]+df['FEPS_2'][i]-df['FEPS_1'][i])/(x**2)-df['prcc_f'][i]
            )
    starting_guess =0.00000000000007
    coe = optimize.fsolve(f_easton,starting_guess)
    COE.append(coe[0])
    F_COE.append(f_easton(coe[0]))
    print("現在跑號碼:",i,"\n實際數值:",coe,"\n function value:",f_easton(coe[0])) 
df['Reaston']=COE
df['Reaston_function_value']=F_COE
```

```python
DES = df.groupby("fyear")['Reaston'].describe().T
display(DES)
```

## Cost of equity  MPEG(Easton) model ---- with imputaiton

```python
COE = [] 
nrow = df.shape[0]
F_COE =[]

for i in range(0,nrow):
    def  f_easton_im(x):
        return (
            (x*df['FDPS_1_im'][i]+df['FEPS_2_im'][i]-df['FEPS_1_im'][i])/(x**2)-df['prcc_f_im'][i]
            )
    starting_guess =0.00000000000007
    coe = optimize.fsolve(f_easton_im,starting_guess)
    COE.append(coe[0])
    F_COE.append(f_easton_im(coe[0]))
    print("現在跑號碼:",i,"\n實際數值:",coe,"\n function value:",f_easton_im(coe[0])) 
df['Reaston_im']=COE
df['Reaston_function_value_im']=F_COE
```

```python
DES = df.groupby("fyear")['Reaston_im'].describe().T
display(DES)
```
