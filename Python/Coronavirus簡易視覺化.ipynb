```
import pandas
import plotly.express as px
```

```
##read the coronavirus dataset from kaggle 
##see the top 10 row of the dataset 
coronavirus=pandas.read_csv("covid_19_data.csv")
coronavirus = coronavirus.drop(["SNo"],axis=1)
coronavirus.head(10)
```

```
coronavirus.groupby(['Country/Region'])['Confirmed', 'Deaths', 'Recovered','Last Update'].max()
```

```
##group by  observationdate、country、province then choose the max vlaue of these variables which are、confirmed、deaths、recovers
#,'Province/State'
#coronavirus['Country_Province'] = coronavirus["Country/Region"]+'_'+coronavirus["Province/State"]	
formated_gdf = coronavirus.groupby(['ObservationDate', 'Country/Region'])['Confirmed', 'Deaths', 'Recovered'].max()
##view the data 
display(formated_gdf.head(5),formated_gdf.tail(5))


##reset the data index
formated_gdf = formated_gdf.reset_index()
##let date convet to date type 
formated_gdf['ObservationDate'] = pandas.to_datetime(formated_gdf['ObservationDate'])
formated_gdf['ObservationDate'] = formated_gdf['ObservationDate'].dt.strftime('%m/%d/%Y')
formated_gdf['size'] = formated_gdf['Confirmed'].pow(0.21)
display(formated_gdf.head(5),formated_gdf.tail(5))
```

```
fig = px.scatter_geo(formated_gdf, locations="Country/Region", locationmode='country names', 
                     color="Confirmed", size='size', hover_name='Country/Region', 
                     range_color= [0, max(formated_gdf['Confirmed'])+2000], 
                     projection="natural earth", animation_frame="ObservationDate", 
                     title='Spread over time')
fig.update(layout_coloraxis_showscale=False)
fig.show()
```
