```python
import re
for i in range(car.shape[0]):
    s = str(car['StationName'].iloc[i])
    result = re.match(r"([\u4e00-\u9fa5]*)([A-Za-z\s]*)", s)
    car.loc[i,'StationNameCH'] =result.group(1) 
    car.loc[i,'StationNameENG'] =result.group(2)

    s = str(car['DestinationStationName'].iloc[i])
    result = re.match(r"([\u4e00-\u9fa5]*)([A-Za-z\s]*)", s)
    car.loc[i,'DestinationStationNameCH'] =result.group(1) 
    car.loc[i,'DestinationStationNameENG'] =result.group(2)
```

```python
car.to_csv('car.csv',encoding='utf-8-sig', index=False)
```
