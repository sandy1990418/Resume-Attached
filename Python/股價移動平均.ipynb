```
!pip install yfinance
import pandas as pd
import plotly.graph_objs as go
import matplotlib.pyplot as plt
import yfinance as yf
import datetime
```

```
#https://towardsdatascience.com/best-5-free-stock-market-apis-in-2019-ad91dddec984

south = yf.Ticker("1702.TW")
south_price = south.history(period="2y")
south_price 
```

```
south_price['Close'].plot()
```

```
##MOVING AVERAGE

south_price["SMA1"] = south_price['Close'].rolling(window=50).mean()
south_price["SMA2"] = south_price['Close'].rolling(window=200).mean()


plt.figure(figsize=(10,10))
plt.plot(south_price['SMA1'], 'g--', label="SMA1")
plt.plot(south_price['SMA2'], 'r--', label="SMA2")
plt.plot(south_price['Close'], label="close")
plt.legend()
plt.show()
```
